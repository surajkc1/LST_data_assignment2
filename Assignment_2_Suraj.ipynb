Land surface temperature is an important environmental parameter. It is different that air temperature and is believed to be more critical factor to assess climate change. The data used in this study is LST of 2010 and LST 0f 2020 for the 20 places. As, trend of change in LST of 2010 and LST of 2020 in the same places has been done, we will be doing paired t test for the analysis to answer the question that, the seen change in LST within 10 years is significant or not.

```python
from scipy.stats.mstats_basic import Ttest_relResult 
df_h= df['SummerLST2020celsius'] #labelling_Data
df_i=df['Summer2010 Celsius']
ttest_rel(df_h, df_i)     #calculating_paired_t_test
```

The test result shows the test statistic "t" is "-4.60" 0.7670 with p-value 0.0021 that means the Land surface temperature has decreased from 2010 to 2020 in the assessed 20 places and the decrement in the temperature is highly statistically significant since p-value is less than 0.05.
